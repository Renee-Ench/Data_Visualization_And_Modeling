import pandas as pd
df = pd.read_csv('ai_dependency_career_anxiety_students.csv')

# Question 1: Average sleep hours
print(df['sleep_hours'].mean())
# The average student is getting about 6.8 hours of sleep, which honestly sounds about right for a busy college schedule. Most of the students in the data are hovering around that 6 or 7 hour mark

# Question 2: Count of ChatGPT as primary tool
chatgpt_users = df[df['primary_ai_tools_used'] == 'ChatGPT'].shape
print(chatgpt_users)
# A ton of people are using ChatGPT as their main tool because it's basically the one everyone knows about. It’s definitely more popular in the survey than other tools like Claude or Perplexity

# Question 3: AI dependency by area
dependency_by_area = df.groupby('urban_or_rural')['ai_dependency_score'].mean()
print(dependency_by_area)
# The dependency scores are actually pretty high for both groups, showing that kids everywhere are leaning on AI for their work. Urban students might have more access to tech, but rural students are using it just as much to keep up

# Question 4: Stress level for "Always" AI users
always_users_stress = df[df['uses_ai_for_assignments'] == 'Always']['stress_level'].mean()
print(always_users_stress)
# Students who always use AI for their homework have pretty high stress levels, which is kind of a mood. Maybe they are stressed because they feel like they aren't actually learning the material themselves

# Question 5: Average fear of job loss by degree type
fear_by_degree = df.groupby('degree_type')['fear_of_job_loss_to_ai'].mean()
print(fear_by_degree)
# Students in technical degrees like B.Tech or B.E. seem to have a lot of fear about AI taking over their future roles. It makes sense because those are the exact jobs that AI is starting to do really well
